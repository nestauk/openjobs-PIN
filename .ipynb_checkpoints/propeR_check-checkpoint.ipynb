```python
# first, load the list of all TTWA
ttwa_data = pd.read_csv(ttwa_file)
# first column is ttwa codes, second column is ttwa names
ttwa_info11 = pd.read_excel(ttwa_info11_file)
ttwa_info16 = pd.read_excel(ttwa_info16_file)
#print(ttwa_info11.tail(n=3))
#print(ttwa_info16.tail(n=3))

# get small TTWAs
small_ttwas = list(ttwa_info11['ttwa11cd'][ttwa_info11['LSOAs']<40])
print('There are {} TTWAs with less than 40 LSOAs.'.format(len(small_ttwas)))

# now set the ttwa code as the index
ttwa_data = ttwa_data.set_index('ttwa11cd')
ttwa_info11 = ttwa_info11.set_index('ttwa11cd')
ttwa_info16 = ttwa_info16.set_index('ttwa11cd')

# drop rows
ttwa_data = ttwa_data.drop(small_ttwas, axis = 0)
ttwa_info11 = ttwa_info11.drop(small_ttwas, axis = 0)
ttwa_info16 = ttwa_info16.drop([t for t in small_ttwas if t in ttwa_info16.index], axis = 0)
ttwa_data = ttwa_data.sort_index()
ttwa_info16 = ttwa_info16.sort_index()
ttwa_info11 = ttwa_info11.sort_index()

# Create aliases for the column names (need to be shorter to be plotted correctly)
rename_cols16 = {'Employment rate ': 'Employment rate',
       '% of economically inactive who want a job':'Job-seeking economically inactive',
       'Claimant Count, % aged 16-64, April 2015 to March 2016 ': 'Claimant count',
       'All in employment who are 1: managers, directors and senior officials (SOC2010)': 'Employed in SOC code 1',
       ' All in employment who are 2: professional occupations or 3: associate prof & tech occupations (SOC2010)': 
                 'Employed in SOC code 2',
       'All in employment who are 5: skilled trades occupations (SOC2010)': 'Employed in SOC code 5',
       'All in employment who are 6: caring, leisure and other service occupations (SOC2010)': 
                 'Employed in SOC code 6',
       'All in employment who are 8: process, plant and machine operatives (SOC2010)':'Employed in SOC code 8',
       'All in employment who are 9: elementary occupations (SOC2010)':'Employed in SOC code 9'}

rename_cols11 = {'Supply-side self-containment (% employed residents who work locally)':
                 'Supply-side self-containment',
       'Demand-side self-containment (% local jobs taken by local residents)':'Demand-side self containment',
       'Number of economically active residents (aged 16+)':'Economically active residents'}
ttwa_info16.rename(rename_cols16, axis = 1, inplace = True)
ttwa_info11.rename(rename_cols11, axis = 1, inplace = True)

ttwa_data = ttwa_data.reset_index()
```

```python
# load the extracted dictionaries of OA centroids
loadOA = True
loadLSOA = True
oa_path = folder4 + 'oa_centroids_dictionary.pickle'
lsoa_path = folder4 + 'lsoa_centroids_dictionary.pickle'
exists = os.path.isfile(oa_path)
if exists and loadOA:
    print('Loading the OA data')
    oa_data = pd.read_pickle(oa_path)
oa_data.rename(columns = {'long': 'lon'}, inplace = True)

exists = os.path.isfile(lsoa_path)
if exists and loadLSOA:
    print('Loading the LSOA data')
    lsoa_data = pd.read_pickle(lsoa_path)
lsoa_data.rename(columns = {'long': 'lon'}, inplace = True)

# Load the data dictionaries which then should be transformed to dataframes and joined.
# They should also be joined with the list of TTWAs for each OA
# Then, I can make the relevant plots
# What I want is a breakdown of mean travel distances for occupations and for ttwa

# first, load the data
with open(save_oa_file, 'rb') as f:
    _,_,oa_occupations,oa_residents,socGroups,_,_ = pickle.load(f)

with open(save_oa_file_jobs, 'rb') as f:
    _,oa_number_of_jobs,oa_jobs_breakdown,jobs_socGroups,_,_ = pickle.load(f)
    
print('Loaded LMIforALL data. Now joining')
t0 = time.time()
# join all dictionaries with oa_data and delete?
# first create the residents column and change the column title
oa_data = oa_data.join(pd.DataFrame.from_dict(oa_residents, orient = 'index'))
# now add everything else
oa_data.rename(columns = {0: 'residents'}, inplace = True)
oa_data = oa_data.join(
    pd.DataFrame.from_dict(oa_occupations, orient = 'index')).join(
    pd.DataFrame.from_dict(oa_number_of_jobs, orient = 'index')).join(
    pd.DataFrame.from_dict(oa_jobs_breakdown, orient = 'index'))
print('It took {:2f}s to create the full dataframe with {} rows'.format(time.time()- t0, len(oa_data)))
# finally, rename the number of jobs column
oa_data.rename(columns = {0: 'residents'}, inplace = True)
print(oa_data.head(n=2))

oa_occupations = None
oa_residents = None
oa_number_of_jobs = None
oa_jobs_breakdown = None
print('Done')
```

```python
# get all the OAs and LSOAs in the Stafford TTWA and call the lmi for all to get the job breakdown
t0 = time.time()
local_ttwa = 'E30000228'
#'E30000228' is for Leamington Spa
#'E30000209' # Gloucester
#'E30000194' Corby
#'E30000189' Cheltenham
#'E30000271' Stafford

local_oa = oa_data[oa_data['ttwa'] == local_ttwa]
local_lsoa = lsoa_data[lsoa_data['ttwa'] == local_ttwa]
ttwa_name = ttwa_data['ttwa11nm'][ttwa_data['ttwa11cd']==local_ttwa].values[0].lower().replace(' ','-') 
#'cheltenham'
print('There are {} OAs in the {} TTWA'.format(len(local_oa), ttwa_name))
print('There are {} LSOAs in the {} TTWA'.format(len(local_lsoa), ttwa_name))
```

```python
# select 5 random origins and destinations OA to get travel times for
ix = np.random.randint(low= 0, high = len(local_oa), size= 10)
origin_oas= local_oa[['lon','lat']].iloc[ix[0:5]].reset_index().rename({'oa11':'name'}, axis = 1)
destination_oas = local_oa[['lon','lat']].iloc[ix[5:]].reset_index().rename({'oa11':'name'}, axis = 1)
# reduce number of digits
origin_oas['lon'] = origin_oas['lon'].map(lambda x: np.around(x,3))
origin_oas['lat'] = origin_oas['lat'].map(lambda x: np.around(x,3))
destination_oas['lon'] = destination_oas['lon'].map(lambda x: np.around(x,3))
destination_oas['lat'] = destination_oas['lat'].map(lambda x: np.around(x,3))
# save them to csv
origin_oas.to_csv(os.path.join(tmp_proper_folder, 'tmp_origin.csv'))
destination_oas.to_csv(os.path.join(tmp_proper_folder, 'tmp_destination.csv'))
# now try and reload them with propeR
origins_df = propeR.importLocationData(os.path.join(tmp_proper_folder, 'tmp_origin.csv'))
destinations_df = propeR.importLocationData(os.path.join(tmp_proper_folder, 'tmp_destination.csv'))
```

```python
# call Transport API for all pairs
trial_journeys_tapi = {}
for ix1 in origin_oas.index:
    origin_oa = origin_oas.loc[ix1]
    trial_journeys_tapi[origin_oa['name']] = {}
    for ix2 in destination_oas.index:
        destination_oa = destination_oas.loc[ix2]
        urlname = urlname_var.format(origin_oa['lon'], origin_oa['lat'], destination_oa['lon'], 
                                     destination_oa['lat']) + urlname_fix
        print(urlname,origin_oa['name'])
        out = requests.get(urlname).json()
        nb_of_calls += 1
        trial_journeys_tapi[origin_oa['name']][destination_oa['name']] = out
```

```python
with open(os.path.join(tmp_proper_folder,'tmp_results_tapi_{}2_car.pickle'.format(ttwa_name)), 'wb') as f:
    pickle.dump(trial_journeys_tapi,f)
```

```python
# check whether propeR and transport API agree

tapi_duration = []
proper_duration1 = []
proper_duration2 = []
for ka in trial_journeys_tapi:
    small_proper1 = proper_res[proper_res['origin'] == ka]
    small_proper2 = proper_res[proper_res['destination'] == ka]
    for kb in trial_journeys_tapi[ka]:
        single_proper1 = small_proper1[small_proper1['destination'] == kb]
        single_proper2 = small_proper2[small_proper2['origin'] == kb]
        tapi_duration.append(trial_journeys_tapi[ka][kb]['routes'][0]['duration'])
        proper_duration1.append(single_proper1['duration_mins'].values)
        proper_duration2.append(single_proper2['duration_mins'].values)
for ix in range(len(tapi_duration)):
    print(tapi_duration[ix],proper_duration1[ix],proper_duration2[ix])
```
