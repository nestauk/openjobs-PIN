```python
def adjust_commute_times(folder3, ttwa_name, dest_lsoas, urlname_var, local_lsoa_ordered, urlname_fix, nb_of_calls,
                         max_calls, CALL_API = False, SAVE_NEW_DICT = False, version_load = '', version_save = '_v2'):
    counter = 0
    for ix in range(len(dest_lsoas)):
        lsoa_commute_file = folder3 + 'PIN/{}_all_lsoa_commute_{}{}.pickle'.format(ttwa_name,dest_lsoas[ix],
                                                                                   version_load)
        # load the file for this LSOA destination
        with open(lsoa_commute_file, 'rb') as f:
            lsoa_commute = pickle.load(f)
        # check if we have data for all origins
        for lsoa in lsoa_commute[dest_lsoas[ix]]:
            # if the routes key is empty, then there was a problem with that API call
            isroutesin = 'routes' in lsoa_commute[dest_lsoas[ix]][lsoa]#['routes']
            if isroutesin:
                isroutesempty = not len(lsoa_commute[dest_lsoas[ix]][lsoa]['routes'])
            else:
                isroutesempty = False
            flag = (not isroutesin) or isroutesempty
            if flag: #not len(lsoa_commute[dest_lsoas[ix]][lsoa]['routes']):
                #break
                # check whether destination and origin coincide
                same_orig_dest = dest_lsoas[ix] == lsoa
                # if they are the same, then set the duration to 0 and leave a note:
                if same_orig_dest:
                    lsoa_commute[dest_lsoas[ix]][lsoa]['routes'] = [
                        {'duration': '00:00:00', 'notes': 'destination = origin'}]
                #    tmp_dict[dest_lsoas[ix]][lsoa]['routes'] = ['destination = origin']
                #    tmp_dict[dest_lsoas[ix]][lsoa]['url_to_call'] = 'none'
                else:
                    print('Missed one because routes not there ({}) or empty ({})'.format(
                        not isroutesin, isroutesempty))
                    print('destination and origin: ',dest_lsoas[ix],lsoa)
                    counter += 1
                    # if destination != origin, then try calling the API again
                    #tmp_dict[dest_lsoas[ix]][lsoa] = {}
                    urlname = urlname_var.format(local_lsoa_ordered['long'].loc[lsoa],
                                         local_lsoa_ordered['lat'].loc[lsoa],
                                         local_lsoa_ordered['long'].loc[dest_lsoas[ix]],
                                         local_lsoa_ordered['lat'].loc[dest_lsoas[ix]]) + urlname_fix
                    print(lsoa_commute[dest_lsoas[ix]][lsoa], urlname)
                    stop
                    #tmp_dict[dest_lsoas[ix]][lsoa]['url_to_call'] = urlname
                    if CALL_API and nb_of_calls<max_calls:
                        out = requests.get(urlname).json()
                        nb_of_calls += 1
                #        tmp_dict[dest_lsoas[ix]][lsoa]['api_result'] = out
                #        print(nb_of_calls)
                        lsoa_commute[dest_lsoas[ix]][lsoa] = out
        if SAVE_NEW_DICT:
            # save commute times file version 2
            lsoa_commute_file_new = folder3 + 'PIN/{}_all_lsoa_commute_{}{}.pickle'.format(ttwa_name,dest_lsoas[ix],
                                                                                           version_save)
            with open(lsoa_commute_file_new, 'wb') as f:
                pickle.dump(lsoa_commute,f)
    flag = counter>0
    # return the updated number of calls            
    return nb_of_calls, flag

# call the function above if we want to adjust the dictionary
ADJUST_TIMES = True
if ADJUST_TIMES:
    nb_of_calls, flag = adjust_commute_times(folder3, ttwa_name, dest_lsoas, urlname_var, local_lsoa_ordered, 
                                             urlname_fix, 
                                       nb_of_calls, max_calls, CALL_API = False, SAVE_NEW_DICT = False, 
                                       version_load = '', version_save = '')
    print('Need to collect some more: {}'.format(flag))
```

```python
def extract_duration(api_call):
    all_durations = []
    for iroute in range(len(api_call['routes'])):
        tmp = pd.DatetimeIndex([api_call['routes'][iroute]['duration']])[0]
        tmp = tmp.hour * 3600 + tmp.minute * 60 + tmp.second
        # now get all the modes of transport used in this route
        if 'route_parts' in api_call['routes'][iroute]:
            all_modes = api_call['routes'][iroute]['route_parts']
            all_modes = [t['mode'] for t in all_modes]
            if 'taxi' in all_modes:
                # if we're getting a taxi, we would have to walk at least three hours
                tmp = 3 * 3600
        # some TTWA have been setup to have zero duration when origin = destination. 
        # So, so not add a null duration to the list
        if tmp >-1:
            all_durations.append(tmp)
    if len(all_durations):
        return min(all_durations)
    else:
        return np.nan

durations_matrix = 1000 + np.zeros((len(dest_lsoas),len(local_lsoa_ordered.index)))
lsoa_rows = list(dest_lsoas)
lsoa_columns = list(local_lsoa_ordered.index)
# now load all the commuting distances again and get the duration
version_load = '' #'_v2' #''
for ix,dest_lsoa in enumerate(dest_lsoas):
    lsoa_commute_file = folder3 + 'PIN/{}_all_lsoa_commute_{}{}.pickle'.format(ttwa_name,dest_lsoa,
                                                                               version_load)
    with open(lsoa_commute_file, 'rb') as f:
        lsoa_commute = pickle.load(f)
    # check if we have data for all origins
    for il,lsoa in enumerate(local_lsoa_ordered.index): #lsoa_commute[dest_lsoa]):
        try:
            durations_matrix[ix,il] = extract_duration(lsoa_commute[dest_lsoa][lsoa])
        except:
            print('Something went wrong when extracting travel duration for ',dest_lsoa, lsoa)
            durations_matrix[ix,il] = np.nan
origin_lsoas = list(local_lsoa_ordered.index)

# now transform it to minutes
durations_matrix = durations_matrix / 60.
print('Done')
# NOTE: If nothing gets printed out, it means that only the diagonal should have NaN value (their 'routes' entry has
# 0 length), because in the cell above I check for missing routes for (destination,origin) pairs where 
# destination!=origin. 
```

```python
# for definitions refer to the google doc
EofC1 = np.sum(Wjt)
with open(os.path.join(plot_save_dir,'out_{}.txt'.format(ttwa_formal_name)), 'w') as f:
    print('Overall, {} has an ease of commute score nb. 1 of {:.3f}'.format(ttwa_formal_name,EofC1), file= f)
    # the following assumes that the origin LSOAs are ordered so that the first D of them are the destinations
    EofC1min = np.sum(population_vector_norm[:len(dest_lsoas)] * jobs_vector_norm)
    EofC1norm = 1/(1-EofC1min) * (EofC1 - EofC1min)
    print('Overall, {} has a normalised ease of commute score nb. 1 of {:.3f}'.format(ttwa_formal_name,EofC1norm),
         file= f)
    m = 0.5
    EofC2 = np.sum(jobs_vector_norm * ((Ajt - m)>0))
    print('Overall, {} has an ease of commute score nb. 2 of {:.1f}%. That is, '.format(ttwa_formal_name,EofC2*100) + 
         'this is the percentage of jobs that can be reached by at least {:.0f}% of the population '.format(100*m) + 
         'within the time constraint', file= f)
# Note: the last sentence means that for each of the jobs in that 58.5% I can find a subset of LSOAs from which it 
# can be reached within 45 minutes that together account for at least 50% of the population
# Saying that no job can be reached by 80% of the population means that there is no specified job that can be reached
# by 80% of the population in less than 45 minutes.

# print again here
print('Overall, {} has an ease of commute score nb. 1 of {:.3f}'.format(ttwa_formal_name,EofC1))
print('Overall, {} has a normalised ease of commute score nb. 1 of {:.3f}'.format(ttwa_formal_name,EofC1norm))
print('Overall, {} has an ease of commute score nb. 2 of {:.1f}%. That is, '.format(ttwa_formal_name,EofC2*100) + 
         'this is the percentage of jobs that can be reached by at least {:.0f}% of the population '.format(100*m) + 
         'within the time constraint')
```

```python
# Compute some needed quantities
base = np.sort(Ajt)
ix_sort = np.argsort(Ajt)
jobs_vector_sorted = jobs_vector_norm[ix_sort]
cumulative = np.arange(1,len(Ajt)+1)/len(Ajt)
cumulative2 = np.cumsum(jobs_vector_sorted)
cumulative2 = cumulative2/np.max(cumulative2)

# compute the EofC2 vector for relevant threshold values
#ms = np.array([0, 0.1, 0.2, 0.325, 0.336, 0.387, 0.419, 0.456,
#        0.494, 0.538, 0.601, 0.612, 0.778, 0.8, 1 ]) #np.arange(0,1,0.05)
ms = np.array([0] + [base[0]-0.001] + [t +.001 for t in list(base[0:-1])] + [base[-1]-.001] + [base[-1]+.001]+ [1])
#np.pad(np.pad(base + .01,((1, 0),), mode = 'constant', constant_values = 0), ((0, 1),), mode = 'constant', 
#           constant_values = 1)# np.array([0, 0.1, 0.2, 0.326, 0.337, 0.388, 0.42, 0.457,
#        0.496, 0.539, 0.603, 0.615, 0.78, 0.9, 1 ])
EofC2_vector = np.sum(jobs_vector_norm[:,np.newaxis] * (np.subtract.outer(Ajt,ms) > 0), axis = 0)
# plot EofC2 vs m
with sns.plotting_context("talk"):
    fig, ax = plt.subplots(figsize=(7,4))
    sns.lineplot(x = ms*100, y = EofC2_vector*100, color='k', marker = 'o')
    sns.lineplot(x = ms*100, y =100 + np.zeros_like(ms), color = [1, 90/255, 0])
    plt.xlabel('population threshold (%)')
    plt.ylabel('percentage of accessible jobs')
    plt.title('TTWA: {}'.format(ttwa_formal_name))
    plt.legend(('Actual curve','Ideal curve'), fontsize= 16, frameon = False)
    # think about the interpretation again +  check R + 
if SAVEFIG:
    plt.savefig(os.path.join(plot_save_dir,'ease_of_commute_score2_distribution_{}.png'.format(ttwa_formal_name)), 
                bbox_inches = 'tight')

with sns.plotting_context('talk'):
    plt.figure(figsize = (7,4))
    plt.plot(base*100,cumulative, marker = 'o')
    plt.plot(ms*100, 1- EofC2_vector, marker = 'o')
    plt.plot(base*100,cumulative2, '--o', markersize = 7)
    plt.legend(['ecdf','1 - EofC2','weighted ecdf'])
    plt.xlabel('population threshold (%)')
    plt.ylabel('probability')
    plt.title('TTWA: {}'.format(ttwa_formal_name))
if SAVEFIG:
    plt.savefig(os.path.join(plot_save_dir,'ease_of_commute_score2_interpretation_{}.png'.format(ttwa_formal_name)), 
                bbox_inches = 'tight')
```
